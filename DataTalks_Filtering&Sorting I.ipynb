# Filtering and Sorting Data
import dataset from https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv

```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep = '\t')
```

```python
df.head(10)
```

#### Reassign the column with the cleaned prices

```python
price = [float(price[1:-1]) for price in df.item_price]
df.item_price = price
```

#### Delete the duplicates in item_name and quantity

```python
df_filter = df.drop_duplicates(["quantity", "item_name", "choice_description"])
quanIs1 = df_filter[df_filter.quantity == 1]
quanIs1
```

#### Filter the item_price > 10

```python
quanIs1[quanIs1["item_price"]>10]
```

#### The price of each item

```python
df[(df["quantity"]==1) & (df["item_name"]=="Steak Bowl")]
```

```python
df[(df["quantity"]==1) & (df["item_name"]=="Chicken Bowl")]
```

#### Select two (or more) columns

```python
df[["item_name", "choice_description", "item_price"]]
```

#### Sort the values from the most to less expensive (DESC)

```python
df.sort_values(by = "item_price", ascending = False)
```

#### Sort by the name of the item

```python
df.sort_values(by = "item_name")
```

#### The quantity of the most expensive item was ordered

```python
df.sort_values(by = "item_price", ascending = False).head(1)
```

#### How many times was a Veggie Salad Bowl ordered?

```python
df[df["item_name"] == "Veggie Salad Bowl"].count()
```

```python
df[df.item_name == "Veggie Salad Bowl"]
```

#### How many times did someone order more than one Canned Soda?

```python
df[(df.item_name == "Canned Soda") & (df.quantity > 1)].count()
```
