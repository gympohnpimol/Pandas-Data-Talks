# Visualization
import dataset from https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style("white")
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv")
```

```python
df.head(10)
```

#### Delete the Unnamed 0 column

```python
del df["Unnamed: 0"]
df
```

#### Plot the total_bill column histogram

```python
hist = sns.distplot(df.total_bill)
hist.set(xlabel = "Value", ylabel = "Frequency", title = "Total Bill")
sns.despine()
```

#### Create a scatter plot presenting the relationship between total_bill and tip

```python
sns.jointplot( x = "total_bill", y = "tip", data = df)
```

#### Create one image with the relationship of total_bill, tip and size

```python
sns.pairplot(df)
```

#### Present the relationship between days and total_bill value

```python
sns.stripplot( x = "day", y = "total_bill", data = df, jitter = True)
```

#### Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.stripplot( x = "tip", y = "day", hue = "sex", data = df, jitter = True)
```

#### Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot( x = "day", y = "total_bill", hue = "time", data = df )
```

#### Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
sns.set(style = "ticks")

g = sns.FacetGrid(df, col = "time")
g.map(plt.hist, "tip")
```

#### Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker They must be side by side.

```python
g = sns.FacetGrid( df, col = "sex", hue = "smoker")
g.map(plt.scatter, "total_bill", "tip", alpha = .7)
g.add_legend()
```
