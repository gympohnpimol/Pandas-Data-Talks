# Deleting
import dataset from https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data')
df
```

#### Create columns for the dataset

```python
df.columns = ['sepal_length','sepal_width', 'petal_length', 'petal_width', 'class']
df
```

#### Is there any missing value in the dataframe?

```python
pd.isnull(df).sum()
```

#### Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN

```python
df.iloc[10:30, 2:3] = np.nan
df.head(30)
```

#### Lets substitute the NaN values to 1.0

```python
df.petal_length.fillna(1, inplace = True)
df.head(20)
```

#### Let's delete the column class

```python
del df['class']
df
```

#### Set the first 3 rows as NaN

```python
df.iloc[0:3,:] = np.nan
df
```

#### Delete the rows that have NaN

```python
df = df.dropna(how='any')
df
```

#### Reset the index so it begins with 0 again

```python
df = df.reset_index(drop=True)
df
```
