import dataset from https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv

```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv")
```

```python
df.head(10)
```

#### Delete the column 'Unnamed: 0' and 'Id'

```python
df = df.drop(["Unnamed: 0", "Id"], axis=1)
df.head()
```

#### Are there more male or female names in the dataset?

```python
df["Gender"].value_counts()
```

#### Group the dataset by name and assign to names

```python
df = df.drop(["Year"], axis =1)
df.head()
```

```python
name = df.groupby("Name").sum()
name.head()
name.sort_values("Count", ascending = 0).head()
```

```python
name.sort_values("Count", ascending = 1).head()
```

#### How many different names exist in the dataset?

```python
len(name)
```

#### What is the name with most occurrences?

```python
name.Count.idxmax()
```

####  How many different names have the least occurrences?

```python
len(name[name.Count == name.Count.min()])
```

#### What is the median name occurrence?

```python
name[name.Count == name.Count.median()]
```

#### What is the standard deviation of names?

```python
name.Count.std()
```

```python
name.Count.median()
```

#### Get a summary with the mean, min, max, std and quartiles

```python
name.describe()
```
