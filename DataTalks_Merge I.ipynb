```python
df1 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv")
df2 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv")
df1.head()
df2.head()
```

```python
df1.head()
```

```python
df1 = df1.loc[:,"mpg":"car"]
df1.head()
```

#### The number of observations in each dataset

```python
print(df1.shape)
print(df2.shape)
```

#### Join df1 and df2 into a single DataFrame called cars

```python
cars = df1.append(df2)
cars
```
